```python
regles = [
    {"conditions": ["fievre", "toux"], "conclusion": "grippe"},                              # R1
    {"conditions": ["grippe", "douleurs_thoraciques"], "conclusion": "infection_respiratoire"}, # R2
    {"conditions": ["infection_respiratoire", "essoufflement"], "conclusion": "hospitalisation_conseillee"}  # R3
]
```

```python
def chainage_avant(faits_initiaux, regles):
    faits = faits_initiaux.copy()  
    nouveau_fait = True

    print("=== Chaînage avant ===")
    print("Faits initiaux :", faits)

    while nouveau_fait:
        nouveau_fait = False
        for regle in regles:
            if all(cond in faits for cond in regle["conditions"]):
                conclusion = regle["conclusion"]
                if conclusion not in faits:
                    faits.append(conclusion)
                    nouveau_fait = True
                    print(f"Règle appliquée : {regle['conditions']} -> {conclusion}")

    print("Faits finaux :", faits)
    print()
    return faits
```

```python
faits_initiaux_1 = ["fievre", "toux"]
print("CAS 1")
faits_finaux_1 = chainage_avant(faits_initiaux_1, regles)
faits_initiaux_2 = ["fievre", "toux", "douleurs_thoraciques", "essoufflement"]
print("CAS 2")
faits_finaux_2 = chainage_avant(faits_initiaux_2, regles)

faits_initiaux_3 = ["fievre"]
print("CAS 3")
faits_finaux_3 = chainage_avant(faits_initiaux_3, regles)
```

```python
def peut_on_deduire(fait_cible, faits_initiaux, regles):
    print(f"=== Peut-on déduire : {fait_cible} ? ===")
    faits_finaux = chainage_avant(faits_initiaux, regles)
    if fait_cible in faits_finaux:
        print(f"Oui, {fait_cible} est déduit par le système.\n")
        return True
    else:
        print(f"Non, le système ne peut pas conclure {fait_cible}.\n")
        return False
```

```python
peut_on_deduire("infection_respiratoire", ["fievre", "toux", "douleurs_thoraciques", "essoufflement"], regles)
peut_on_deduire("infection_respiratoire", ["fievre", "toux"], regles)
```
